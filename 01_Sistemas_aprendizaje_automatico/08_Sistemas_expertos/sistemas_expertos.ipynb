## üìã Resumen

Sistemas Expertos capturan conocimiento humano experto mediante reglas y realizan razonamiento automatizado.

### üéØ Objetivos
- **Representaci√≥n**: Reglas, frames, redes sem√°nticas, ontolog√≠as
- **Inferencia**: Forward chaining, backward chaining
- **L√≥gica Difusa**: Razonamiento con incertidumbre
- **Motores de Reglas**: IF-THEN rules
- **Explicabilidad**: Justificar conclusiones

### üéõÔ∏è Componentes

1. **Base de Conocimiento**: Reglas + Hechos
2. **Motor de Inferencia**: Forward/Backward chaining
3. **Interfaz de Usuario**: Interacci√≥n
4. **M√≥dulo de Explicaci√≥n**: Justificaci√≥n

### üè• Sistemas Cl√°sicos
- **MYCIN**: Diagn√≥stico de infecciones bacterianas
- **DENDRAL**: Identificaci√≥n de estructuras moleculares
- **XCON/R1**: Configuraci√≥n de sistemas inform√°ticos
- **CLIPS**: Lenguaje de programaci√≥n de sistemas expertos

---

```python
# Setup - Sistema Experto Simple
class ExpertSystem:
    def __init__(self):
        self.facts = set()
        self.rules = []
    
    def add_fact(self, fact):
        self.facts.add(fact)
    
    def add_rule(self, conditions, conclusion):
        self.rules.append((conditions, conclusion))
    
    def forward_chaining(self):
        inferred = set()
        for conditions, conclusion in self.rules:
            if all(c in self.facts for c in conditions):
                inferred.add(conclusion)
        return inferred

# Ejemplo: Diagn√≥stico m√©dico simple
es = ExpertSystem()
es.add_fact('fiebre')
es.add_fact('tos')
es.add_rule(['fiebre', 'tos'], 'gripe')
es.add_rule(['fiebre'], 'infecci√≥n')

diagnosticos = es.forward_chaining()
print(f"S√≠ntomas: {es.facts}")
print(f"Diagn√≥sticos: {diagnosticos}")
print("\n‚úÖ Sistema Experto b√°sico implementado")
print("\nüìö M√©todos de Inferencia:")
print("  ‚Ä¢ Forward Chaining: De datos ‚Üí conclusiones")
print("  ‚Ä¢ Backward Chaining: De objetivos ‚Üí datos")
print("\nüîë Aplicaciones:")
print("  ‚Ä¢ Diagn√≥stico m√©dico, Configuraci√≥n, Planificaci√≥n")
```
